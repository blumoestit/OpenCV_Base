```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("Blooms.jpg")
img
```

### OpenCV has a different channel order -> BGR, not RGB as in Matplotlib

```python
plt.imshow(img);
```

### Convert to right rgb colors

```python
# Covert BGR to RGB - there are many different types of convert
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 
```

```python
img_rgb.shape
```

```python
plt.imshow(img_rgb);
```

### Load directly image in gray

```python
img_gray = cv2.imread('Blooms.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray, cmap='gray');
```

```python
img_gray.shape
```

### Resize image

```python
img = cv2.resize(img_rgb, (1500, 1000))
plt.imshow(img);
```

#### Caution -> hight and width change order in shape

```python
img.shape
```

### Use ratios

```python
w_ratio = 0.2
h_ratio = 0.5
```

```python
img_ratio = cv2.resize(img_rgb,(0,0),img,w_ratio,h_ratio)
```

```python
plt.imshow(img_ratio);
```

### Flip image

```python
img_flip = cv2.flip(img_ratio, 0) # 0 means x axis, 1 y axis
plt.imshow(img_flip);
```

### Save the changed images

```python
#cv2.imwrite('new_blooms.jpg', img_flip)
```

### Change image size in notebook

```python
fig = plt.figure(figsize=(15,12))
ax = fig.add_subplot(111)
ax.imshow(img_flip);
```
