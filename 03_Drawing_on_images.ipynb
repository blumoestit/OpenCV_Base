```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
blank_img = np.zeros(shape=(512, 512, 3), dtype=np.int16)
```

```python
plt.imshow(blank_img);
```

### Shapes

```python
# pt1 = up left
# pt2 = down right
#pt1 = (200, 300)
#pt2 = (400, 400)
#color = (0, 0, 255)
cv2.rectangle(blank_img, 
              pt1=(50, 100), 
              pt2=(450, 350), 
              color=(255, 255, 255),
              thickness=5)
plt.imshow(blank_img);
```

```python
cv2.circle(blank_img, 
           center=(250, 225),
           radius=100,
           color=(255, 20, 40),
           thickness=-1)
plt.imshow(blank_img);
```

### Text

```python
font = cv2.FONT_HERSHEY_DUPLEX 
cv2.putText(blank_img,
           text = 'Japan',
           org=(50,450),
           fontFace=font,
           fontScale = 4,
           color=(250, 250, 250))
plt.imshow(blank_img);
```

### Polygons

```python
# numer of rows = numer of the vertices
polygon_img = np.zeros(shape=(512,512,3),dtype=np.int32)
```

```python
vertices = np.array([[100,300],
                     [200,200],
                     [400,300],
                     [200,400]],
                     np.int32)
```

```python
pts = vertices.reshape((-1,1,2))
```

```python
pts
```

```python
cv2.polylines(polygon_img,[pts],
              isClosed=True,
              color=(255,0,150),
              thickness=5)
plt.imshow(polygon_img)
```
